```python
import csv # csv읽기, 쓰기
import folium # 지도생성

# 파일 읽기 (euc-kr)
file = open('csv/부산광역시_해운대구_도시공원_20201215_1608034363405_11422.csv', 'r', encoding="euc-kr")
rdr = csv.reader(file)

next(rdr, None) # 컬럼명 skip

# 읽은 파일 내용 보기 (공원명[1], 위도[5], 경도[6])
for idx, tmp in enumerate(rdr):
    if idx == 0: # 최초 1번 객체 만들기 시작 주소의 중심이 location에 정의된 곳
        map_data = folium.Map(location=[ float(tmp[5]), float(tmp[6]) ], zoom_size=12, zoom_start=12)
    
    # 71개
    folium.Marker([ float(tmp[5]), float(tmp[6]) ], popup=tmp[1]).add_to(map_data)
    
    print(idx, tmp[1], tmp[5], tmp[6])

map_data.save('map/ex01.html') # mpa폴더 생성 지도를 html 저장
```

```python
import csv
import requests
import pandas as pd

# 데이터 1000개 -> 800개 훈련, 200개 테스트 데이터
csv_url = "http://ihongss.com/resources/train.csv"

with requests.Session() as s :
    download = s.get(csv_url)
    decode_content = download.content.decode('utf-8')
    
    file = csv.reader(decode_content.splitlines(), delimiter=",")
    
    data = list(file) #[ [컬럼명], [실제 데이터], [실제 데이터] .... [] ]
#     for tmp in data:
#         print(tmp)

    # 데이터 프레임 만들기 (실제 데이터, 컬럼 데이터)
    df = pd.DataFrame( data[1:], columns=data[0] )
df

df.info()
df.describe()
```
