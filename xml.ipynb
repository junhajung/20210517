```python
#  $eq  (equals) 주어진 값과 일치하는 값 
#  $gt  (greater than) 주어진 값보다 큰 값 
#  $gte  (greater than or equals) 주어진 값보다 크거나 같은 값 
#  $lt  (less than) 주어진 값보다 작은 값 
#  $lte  (less then or equals) 주어진 값보다 작거나 같은 값 
#  $ne  (not equlas) 주어진 값과 일치하지 않는 값 
#  $in  (in) 주어진 배열 안에 속하는 값
#  $nin  (not in) 주어진 배열 안에 속하지 않는 값 

# age가 50보다 큰것 가져오기
data = collection.find({'age':{'$gt':50}},{'_id':0})
for tmp in data:
    print(tmp, type(tmp))
    
```

```python
# 이름, 나이, 주소속성, 주소값 => 가져와서 => [{ },{ }] => DB에 추가
import urllib.request as RQ
import xml.etree.ElementTree as ET

url = "http://ihongss.com/resources/student.xml"

# 웹에서 데이터를 받음
response = RQ.urlopen(url).read()

# 타입 변경 (bytes -> xml.etree.ElementTree.Element)
root = ET.fromstring(response)

arr = []
# for tmp in root.iterfind("student"):  #아래 for문과 동일
for tmp in root.findall("student"):
    print(tmp)
    # 하위 name태의 값, age값, addr의 속성값(딕셔너리), addr의 값
    print(tmp.find("name").text, 
          tmp.findtext("age"),
          tmp.find("addr").attrib['id'],
          tmp.findtext("addr"))

    tmp1 = {"name" : tmp.find("name").text, "age" : tmp.findtext("age"), "addr_id" : tmp.find("addr").attrib['id'], "addr" : tmp.findtext("addr") }
    arr.append(tmp1)

print(arr)
```

```python
# http://webdam.inria.fr/Jorge/files/movies.xml
# 수집항목 : 제목, 개봉일, 나라, 장르, 감독이름, 감독출생일, 배우이름, 출생일, 역할 (여러명)

import urllib.request as RQ
import xml.etree.ElementTree as ET

url = "http://webdam.inria.fr/Jorge/files/movies.xml"

# 웹에서 데이터를 받음
response = RQ.urlopen(url).read()

# 타입 변경 (bytes -> xml.etree.ElementTree.Element)
root = ET.fromstring(response)

for tmp in root:
    print(tmp.find("title").text, tmp.findtext("year"))
    for tmp1 in tmp.findall('director'):
        print("감독 : " , tmp1.find("last_name").text, tmp1.find("first_name").text, tmp1.findtext("birth_date"))
    
    for tmp1 in tmp.findall('actor'):
        print("배우 : ", tmp1.find("last_name").text)
    
    print("=" * 30)
    
```

```python
#리스트에 딕셔너리 타입으로 변경하기
import urllib.request as RQ
import xml.etree.ElementTree as ET

url = "http://webdam.inria.fr/Jorge/files/movies.xml"

# 웹에서 데이터를 받음
response = RQ.urlopen(url).read()

# 타입 변경 (bytes -> xml.etree.ElementTree.Element)
root = ET.fromstring(response)

# [{   },{    },{    }]
arr = []
for tmp in root:
    #딕셔너리 생성
    data = {"title" : tmp.find("title").text, "year" : tmp.findtext("year") }

    for tmp1 in tmp.findall('director'):
        #딕셔너리에 키가 다른 값 추가
        data['director_name'] = tmp1.find("last_name").text + " " + tmp1.find("first_name").text
        data['director_birth_date'] = int(tmp1.findtext("birth_date"))
    
    # enumerate => 반복시  idx번호, tmp1값
    for idx, tmp1 in enumerate(tmp.findall('actor')):
        data['actor_name_' + str(idx)] = tmp1.find("last_name").text + " " + tmp1.find("first_name").text
        data['actor_birth_date_' + str(idx)] = tmp1.find("birth_date").text
        data['actor_role_' + str(idx)] = tmp1.find("role").text
    
    arr.append(data);

print(arr)

```

```python
# http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109
# 지역(province), 도시(city), 시간(tmEf), 정보(wf), 최소온도(tmn), 최고온도(tmx), 비올확인(tnSt)
# weather_20210520 컬렉션에 보관하기

import urllib.request as RQ
import xml.etree.ElementTree as ET

url = "http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"

# 웹에서 데이터를 받음
response = RQ.urlopen(url).read()

# 타입 변경 (bytes -> xml.etree.ElementTree.Element)
root = ET.fromstring(response)

# [{   },{    },{    }]
arr = []
for idx, tmp in enumerate(root.findall('channel/item/description/body/location')):
    for tmp1 in tmp.findall('data'):
        dic = { 'province': tmp.findtext('province'), 
                'city'    : tmp.findtext('city'),
                'tmEf'    : tmp1.findtext('tmEf'),
                'wf'      : tmp1.findtext('wf'),
                'tmn'     : tmp1.findtext('tmn'),
                'tmx'     : tmp1.findtext('tmx'),
                'rnSt'    : tmp1.findtext('rnSt') }
        arr.append(dic)
        
print(len(arr))
arr #print(arr)
```
