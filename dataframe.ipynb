```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'

res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

arr = []
for tmp in soup.select('location'):
    for tmp1 in tmp.select("data"):
#         print(tmp.select_one('city').text, end=" ")
#         print(tmp.select_one('tmEf').text, end=" ")
#         print(tmp.select_one('wf').text, end=" ")
#         print(tmp.select_one('tmn').text, end=" ")
#         print(tmp.select_one('tmx').text, end=" ")
#         print(tmp.select_one('rnSt').text)
#         print("=" * 30)
        arr.append([tmp.select_one('city').text,
                    tmp.select_one('tmEf').text,
                    tmp.select_one('wf').text,
                    tmp.select_one('tmn').text,
                    tmp.select_one('tmx').text,
                    tmp.select_one('rnSt').text] )
df4 = pd.DataFrame(arr, columns=["city","tmef","wf","tmn","tmx","rnSt"])
df4

df4.dtypes
```

```python
df4 = df4.astype({"tmef" : "datetime64", "tmn" : "int64", "tmx":"int64", "rnSt" : "int64"})
```
